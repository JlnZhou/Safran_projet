# Active contours method

Basic idea of active contours method is that finding the best curve along which the integration of $g(x, y)$ function is minimized, where 

$$g(x, y) = \frac{1}{1+\nabla (G *I)}$$

The problem can be reformulated as:

$$\text{desired contour} = \arg\min \int_{\text{curve}} g(x, y) ds$$

We use active contours method to detect fibers shown below. The results is appealing; however, to detect each fiber, we need to specify some initial contour and determine several parameters to get the better result. 

```python
from skimage.external import tifffile
from skimage import img_as_float
from skimage.segmentation import active_contour
from skimage.filters import gaussian
from skimage.filters import threshold_otsu
import numpy as np
import matplotlib.pyplot as plt

img = tifffile.imread("../test_image.tif")
img = img_as_float(img)
th_otsu = threshold_otsu(img)
img = img < th_otsu
```

```python
# 1st fiber
s = np.linspace(0, 2*np.pi, 400)
x = 100 + 100*np.cos(s)
y = 140 + 50*np.sin(s)
init = np.array([x, y]).T
snake1 = active_contour(gaussian(img, 3),
                        init, alpha=0.05, beta=20, gamma=0.001)
# 2nd fiber
s = np.linspace(0, 2*np.pi, 400)
x = 150 + 100*np.cos(s)
y = 140 + 50*np.sin(s)
init = np.array([x, y]).T
snake2 = active_contour(gaussian(img, 3),
                        init, alpha=0.05, beta=20, gamma=0.001)
# 3rd filter
s = np.linspace(0, 2*np.pi, 400)
x = 300 + 100*np.cos(s)
y = 140 + 50*np.sin(s)
init = np.array([x, y]).T
snake3 = active_contour(gaussian(img, 3),
                       init, alpha=0.05, beta=20, gamma=0.001)
# 4th filter
s = np.linspace(0, 2*np.pi, 400)
x = 440 + 100*np.cos(s)
y = 140 + 50*np.sin(s)
init = np.array([x, y]).T
snake4 = active_contour(gaussian(img, 3),
                        init, alpha=0.05, beta=20, gamma=0.001)
# 5th filter
s = np.linspace(0, 2*np.pi, 400)
x = 580 + 100*np.cos(s)
y = 140 + 50*np.sin(s)
init = np.array([x, y]).T
snake5 = active_contour(gaussian(img, 3),
                        init, alpha=0.05, beta=20, gamma=0.001)

%matplotlib inline
fig, ax = plt.subplots(figsize=(15, 8))
ax.imshow(img, cmap=plt.cm.gray)
#ax.plot(init[:, 0], init[:, 1], '--r', lw=3)
ax.plot(snake1[:, 0], snake1[:, 1], '-b', lw=3)
ax.plot(snake2[:, 0], snake2[:, 1], '-b', lw=3)
ax.plot(snake3[:, 0], snake3[:, 1], '-b', lw=3)
ax.plot(snake4[:, 0], snake4[:, 1], '-b', lw=3)
ax.plot(snake5[:, 0], snake5[:, 1], '-b', lw=3)
ax.axis([0, img.shape[1], img.shape[0], 0])
plt.show()
```
